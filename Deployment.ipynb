```python
import pickle
```

```python
import pandas as pd
```

```python
loaded_model=pickle.load(open("Predicting_Calories_Burned_During_Workout.sav",'rb'))
```

```python
session=int(input("Enter the no.of hours spent in gym daily:"))
```

```python
fat=int(input("Enter the fat percentage of your body:"))
```

```python
water=int(input("Enter the amount of water you intake daily(in litres):"))
```

```python
day=int(input("Enter the no.of days you visit gym in a week:"))
```

```python
exp=int(input("Enter the experience level you have in gym(Rate 1 to 5):"))
```

```python
new_data = pd.DataFrame([[session,fat,water,day,exp]], 
                        columns=['Session_Duration (hours)',
                                 'Fat_Percentage',
                                 'Water_Intake (liters)',
                                 'Workout_Frequency (days/week)',
                                 'Experience_Level'])
prediction = loaded_model.predict(new_data)
```

```python
print("The amount of Calories burned is:",prediction[0])
```
